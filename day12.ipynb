```python
import matplotlib.pyplot as plt
import numpy as np



def make_graph(x,y):
        
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x_minus_x_mean = x -x_mean
    y_minus_y_mean = y -y_mean
    contributions = x_minus_x_mean * y_minus_y_mean
    np.cov(x,y)
    plt.scatter(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('covarience')
    plt.axhline(y_mean,linestyle = ':',c='r',label='y_mean')
    plt.axhline(x_mean,linestyle =':', c='g',label='x_mean')
    for xi, yi in zip(x,y):
        plt.plot([xi,x_mean],[yi,yi])
    for xi, yi in zip(x,y):
        plt.plot([xi,xi],[yi,y_mean])
    for xi, yi, contributioni in zip(x,y, contributions):
        plt.text(xi, yi, f"{(xi-x_mean)}*{(yi-y_mean).round(2)} = {contributioni.round(2)}")
    plt.grid(True)
    plt.legend()


#to fix indentation for function ctrl + [ or ]
```

```python
x = np.array([1,2,3,4,5,6,7])
y_positive = np.array([2,3,5,4,6,8,9])
y_negative = -y_positive
print("Positive y values:", y_positive)
print("Negative y values:", y_negative)
make_graph(x, y_positive)
```
