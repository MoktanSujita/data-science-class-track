```python
import numpy as np

y_true = np.array([21,20.1,22.8,55,19.34])
y_pred = np.array([21.5,22.5,7.32,12,20.53])
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true))**2)
r_squared = 1 - (ss_res / ss_tot)

n = len(y_true)
p = 3

adj_r_squared = 1- (1 - r_squared) * ((n-1)/ (n - p -1))
r_squared
adj_r_squared
```

```python

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Calculate the correlation matrix
corr_matrix = np.corrcoef(x, y)
print(corr_matrix)
# Output: [[1. 1.] [1. 1.]]

# Extract the correlation coefficient
r = corr_matrix[0, 1]  # or corr_matrix[1, 0]
print(r) # Output: 1.0   
```

```python
def ks_shift(expected, actual):
    vals = np.sort(np.unique(np.concatenate([expected, actual])))
    cdf_e = np.searchsorted(np.sort(expected), vals, side='right')/len(expected)
    cdf_a = np.searchsorted(np.sort(actual), vals, side='right')/len(actual)
    return np.max(np.abs(cdf_e - cdf_a))

train_data = np.random.normal(0, 1, 1000)
test_data = np.random.normal(0.5, 1.2, 1000)
ks_stat = ks_shift(train_data, test_data)
ks_stat
```
