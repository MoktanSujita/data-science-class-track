```python
x = np.arange(0,5,.1)
y = np.sin(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sin(x)')
plt.plot(x,y,'g')
```

```python
x = np.arange(0,5,.1)
y = np.cos(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=cos(x)')
plt.plot(x,y,'g')
```

```python
#oop
x = np.arange(0,5,.1)
y = np.cos(x)
fig ,ax = plt.subplots()
ax.plot(x,y,'g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('y=cos(x)')
ax.grid(True)
```
