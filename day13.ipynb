```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data_df=pd.read_csv("Advertising.csv")
features=["TV","radio","newspaper"]
colors=["red","green","blue"]
fig,axes=plt.subplots(1,3,figsize=(15,4))
for ax,feature,color in zip(axes,features,colors):
    ax.scatter(data_df[feature],data_df["sales"],color=color,marker=".")
    ax.set_xlabel(features)
    ax.set_ylabel("sales")
    ax.set_title(f"sales vs{features}")
plt.tight_layout()
plt.show()
```

```python
experience=np.array([1,2,3,4,5])
actual_salary=np.array([3000,3500,5000,5500,6500])
b=2000
m_value=np.arange(500,1300,50)
costs=[]
for m in m_value:
    predicted_salary=m*experience + b
    mse=np.mean((actual_salary-predicted_salary)**2)
    costs.append(mse)
    print (costs)
    plt.xlabel('cost')
    plt.ylabel('Mvalue')
plt.plot(m_value,costs)
```
