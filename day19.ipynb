```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10,10,100)
print(x)

def sigmoid(x):
   return 1/(1+np.exp(-x))

y = sigmoid(x)
y
plt.plot(x,y)
plt.title("sigmoid function")
plt.ylabel("sigmoid(x)")
plt.xlabel("x")
#[0-1] function is used for probability functions
```

```python
plt.plot(x,y, linewidth = 2, label = 'Sigmoid Curve')
plt.axhline(y=0.5, linestyle = '--',label = 'y=0.5')
plt.axvline(x=0, linestyle= '--', label = 'x=0')
plt.title("sigmoid Function graph")
plt.xlabel('x')
plt.ylabel('sigmoid')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

y_pred = np.linspace(0.01, 0.99, 100)
y_pred
```

```python
loss_y1 = -np.log(y_pred)
loss_y0 = -np.log(1-y_pred)
loss_y1, loss_y0
```

```python
#binary cross Entropy -> loss function
# why use binary cross entropy
plt.plot(y_pred, loss_y1, label = "Loss when y=1")
plt.plot(y_pred, loss_y0, label = "Loss when y=0")
plt.title("logistic Regression Loss Function")
plt.xlabel("Predicted Probability")
plt.ylabel("Loss")
plt.axvline(x=0.5, linestyle = '--')
plt.grid(True)
plt.legend()
plt.show() #unless it is written function is also shown of matplotlib
```
