```python
x_mean=np.mean(x)
y_mean=np.mean(y)
n=len(x)
np.sum((x-x_mean)*(y-y_mean))/(n-1)
print(f"x-x_mean -> {x-x_mean}) \n y-y_mean ->{(y-y_mean).round(2)}")
print(f"((x-x_mean)*(y-y_mean) -> {(x-x_mean)*(y-y_mean).round(2)})")
print(f"(sum of (x-x_mean)*(y-y_mean)(n -1) -> {(np.sum((x-x_mean)*(y-y_mean)).round(2))/(n-1)}")
```

```python
cov_xx = np.sum((x-x_mean)*(x-x_mean)).round(2)/(n-1)
cov_xy = np.sum((x-x_mean)*(y-y_mean)).round(2)/(n-1)
cov_yx = np.sum((y-y_mean)*(x-x_mean)).round(2)/(n-1)
cov_yy = np.sum((y-y_mean)*(y-y_mean)).round(2)/(n-1)
np.array([cov_xx, cov_xy, cov_yx, cov_yy]).reshape(2,2)
```

```python
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5,6,7])
y = np.array([2,3,5,4,6,8,9])
x_minus_x_mean = x -x_mean
y_minus_y_mean = y -y_mean
contributions = x_minus_x_mean * y_minus_y_mean
np.cov(x,y)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('covarience')
plt.axhline(y_mean,linestyle = '--',c='r',label='y_mean')
plt.axhline(x_mean,linestyle ='--', c='g',label='x_mean')
for xi, yi in zip(x,y):
    plt.plot([xi,x_mean],[yi,yi])
for xi, yi in zip(x,y):
    plt.plot([xi,xi],[yi,y_mean])
for xi, yi, contributioni in zip(x,y, contributions):
    plt.text(xi, yi, f"{(xi-x_mean)}*{(yi-y_mean).round(2)}")
plt.grid(True)
plt.legend()
```
